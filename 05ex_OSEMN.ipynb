1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
%matplotlib inline
import pandas as pd
import numpy as np
import numpy.random as npr

rand_list = npr.rand(10)
txtname = "simple_data.txt"
with open(txtname, mode='w') as f: #a if I want to append data
    # print the list in it
    counter = 0
    for number in rand_list:
        f.write(f"{counter}, {number}\n")
        counter += 1
    
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
import pandas as pd
import numpy as np
import numpy.random as npr

rand_matrix = npr.rand(5,5)
txtname = "data.txt"

with open(txtname, mode = 'w') as f:
    for i in range(5):
        if (i>0): f.write("\n")
        for j in range(5):
            f.write(f"{rand_matrix[i][j]} ")
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
import csv

txtname = "data.txt"
csvname = "data.csv"

with open(txtname, 'r') as ifile, open(csvname, 'w') as ofile:
    for line in ifile:    
        data_str = line.strip().replace(' ',',')
        ofile.write(data_str + "\n")

print(pd.read_csv(csvname))
```
