4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
cards_numbers = []
with open("credit_card.dat", 'r') as file:
    for line in file:
        number = ""
        if (len(line)-5 > 0): 
            for i in range(0, len(line)-5, 6):
                number += chr(int(line[i:i+6], 2)) #The chr function converts integers into characters, which can still be numbers. Check the ASCII table
            cards_numbers.append(number)
print(cards_numbers)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
import csv

userdata = json.load(open('user_data.json'))

filtered = [item for item in userdata if (item['CreditCardType'] == 'American Express')] #This is a list of dictioraries

with open("american_express.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    # Add head line
    writer.writerow(["ID", "Name", "Email", "CreditCard", "CreditCardType"])
    #Add actual data
    for item in filtered:
        writer.writerow([
            item["ID"],
            item["FirstNameLastName"],
            item["EmailAddress"],
            item["CreditCard"],
            item["CreditCardType"]
        ])
```
